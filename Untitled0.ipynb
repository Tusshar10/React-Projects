```python
# Import required libraries
import pandas as pd
import numpy as np
import cupy as cp

# Load the WISDM dataset
df = pd.read_csv(
    "/content/drive/MyDrive/WISDM_ar_v1.1_raw.txt",
    sep=",",
    lineterminator=";",
    header=None,
    names=["user", "activity", "timestamp", "x", "y", "z"],
    on_bad_lines="skip"
)

# Drop NaN values
df = df.dropna()
df['timestamp'] = df['timestamp'] / 1e9  # Convert nanoseconds to seconds

# Normalize timestamps per user
df['timestamp'] = df.groupby('user')['timestamp'].transform(lambda x: x - x.min())

# Original and target sampling rates
old_rate = 20  # Hz
new_rate = 26  # Hz

# Custom linear interpolation function using CuPy
def custom_gpu_interp(x_old, y_old, x_new):
    # Implementation of linear interpolation using CuPy
    # Find indices of x_old that bound each value in x_new
    indices = cp.searchsorted(x_old, x_new)

    # Handle edge cases
    indices = cp.clip(indices, 1, len(x_old) - 1)

    # Get left and right boundary points
    x_left = x_old[indices - 1]
    x_right = x_old[indices]
    y_left = y_old[indices - 1]
    y_right = y_old[indices]

    # Calculate interpolation weights
    weights = (x_new - x_left) / (x_right - x_left)

    # Perform linear interpolation
    return y_left + weights * (y_right - y_left)

# Function to interpolate data for each user using GPU
def interpolate_user_data(user_df):
    # Sort by timestamp and remove duplicates
    user_df = user_df.sort_values(by='timestamp').drop_duplicates(subset=['timestamp'])

    # Convert data to CuPy arrays for GPU processing
    old_timestamps = cp.array(user_df['timestamp'].values)
    x = cp.array(user_df['x'].values)
    y = cp.array(user_df['y'].values)
    z = cp.array(user_df['z'].values)
    activities = user_df['activity'].values  # Keep on CPU for simplicity

    # Generate new timestamps on GPU
    new_timestamps = cp.arange(old_timestamps[0], old_timestamps[-1], 1/new_rate)

    # Perform interpolation on GPU using our custom function
    x_new = custom_gpu_interp(old_timestamps, x, new_timestamps)
    y_new = custom_gpu_interp(old_timestamps, y, new_timestamps)
    z_new = custom_gpu_interp(old_timestamps, z, new_timestamps)

    # Handle activity assignment on CPU
    old_timestamps_np = cp.asnumpy(old_timestamps)
    new_timestamps_np = cp.asnumpy(new_timestamps)
    activity_indices = np.searchsorted(old_timestamps_np, new_timestamps_np, side='right') - 1
    activity_indices = np.clip(activity_indices, 0, len(activities) - 1)
    activity_values = activities[activity_indices]

    # Return DataFrame with results converted back to NumPy arrays
    return pd.DataFrame({
        'user': user_df['user'].iloc[0],
        'activity': activity_values,
        'timestamp': cp.asnumpy(new_timestamps),
        'x': cp.asnumpy(x_new),
        'y': cp.asnumpy(y_new),
        'z': cp.asnumpy(z_new)
    })

# Apply interpolation per user
upsampled_df = df.groupby('user', group_keys=False).apply(interpolate_user_data)

# Function to keep the first 10% of data per activity (remains on CPU)
def keep_first_10_percent(activity_df):
    activity_df = activity_df.sort_values(by="timestamp")
    first_10_percent = activity_df.iloc[: int(len(activity_df) * 0.1)]
    return first_10_percent

# Apply data reduction per activity
reduced_df = upsampled_df.groupby("activity", group_keys=False).apply(keep_first_10_percent)

# Save the reduced dataset
reduced_df.to_csv("WISDM_Upsampled_26Hz_Reduced.csv", index=False)

print("Upsampling complete. Dropped last 90% of data per activity. New dataset saved.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from torch.utils.data import DataLoader, TensorDataset

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load and prepare data
df = pd.read_csv("/content/WISDM_Upsampled_26Hz_Reduced.csv")
activities = df['activity'].unique()

# Parameters
SEQ_LENGTH = 128  # 5 seconds at 26Hz
BATCH_SIZE = 256
LATENT_DIM = 100
NUM_EPOCHS = 20
LEARNING_RATE = 0.0002

# Preprocessing
# Encode labels
le = LabelEncoder()
ohe = OneHotEncoder(sparse_output=False)
df['activity_code'] = le.fit_transform(df['activity'])
activity_labels = ohe.fit_transform(df[['activity_code']])

# Normalize data to [-1, 1]
sensor_data = df[['x', 'y', 'z']].values
sensor_data = 2 * (sensor_data - (-4)) / (4 - (-4)) - 1  # Scale to [-1, 1]

# Create sequences
def create_sequences(data, labels, seq_length):
    sequences = []
    labels_seq = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
        labels_seq.append(labels[i+seq_length-1])  # Use last label in sequence
    return np.array(sequences), np.array(labels_seq)

X, y = create_sequences(sensor_data, activity_labels, SEQ_LENGTH)

# Convert to tensors
X_tensor = torch.FloatTensor(X).to(device)
y_tensor = torch.FloatTensor(y).to(device)

```

# New section

```python
# Create dataset and dataloader
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# CGAN Model Definitions
class Generator(nn.Module):
    def __init__(self, latent_dim, num_classes, output_dim):
        super(Generator, self).__init__()
        self.num_classes = num_classes
        self.lstm = nn.LSTM(latent_dim + num_classes, 128, num_layers=2, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(128, 64),
            nn.LeakyReLU(0.2),
            nn.Linear(64, output_dim),
            nn.Tanh()
        )

    def forward(self, noise, labels):
        seq_len = noise.size(1)
        labels = labels.unsqueeze(1).repeat(1, seq_len, 1)
        combined = torch.cat([noise, labels], dim=-1)
        lstm_out, _ = self.lstm(combined)
        output = self.fc(lstm_out)
        return output

class Discriminator(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(Discriminator, self).__init__()
        self.num_classes = num_classes
        self.lstm = nn.LSTM(input_dim + num_classes, 128, num_layers=2, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(128, 64),
            nn.LeakyReLU(0.2),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, inputs, labels):
        seq_len = inputs.size(1)
        labels = labels.unsqueeze(1).repeat(1, seq_len, 1)
        combined = torch.cat([inputs, labels], dim=-1)
        lstm_out, _ = self.lstm(combined)
        validity = self.fc(lstm_out[:, -1, :])  # Use last timestep
        return validity

# Initialize models
generator = Generator(LATENT_DIM, len(activities), 3).to(device)
discriminator = Discriminator(3, len(activities)).to(device)

# Optimizers
optimizer_G = optim.Adam(generator.parameters(), lr=LEARNING_RATE)
optimizer_D = optim.Adam(discriminator.parameters(), lr=LEARNING_RATE)

# Loss function
adversarial_loss = nn.BCELoss()

# Training loop
for epoch in range(NUM_EPOCHS):
    correct_real = 0
    correct_fake = 0
    total_samples = 0

    for i, (real_data, real_labels) in enumerate(dataloader):
        batch_size = real_data.size(0)
        total_samples += batch_size

        # Adversarial ground truths
        valid = torch.ones((batch_size, 1), device=device)
        fake = torch.zeros((batch_size, 1), device=device)

        # ---------------------
        #  Train Discriminator
        # ---------------------
        optimizer_D.zero_grad()

        # Real data
        real_validity = discriminator(real_data, real_labels)
        d_real_loss = adversarial_loss(real_validity, valid)
        correct_real += ((real_validity > 0.5).float() == valid).sum().item()

        # Fake data
        noise = torch.randn((batch_size, SEQ_LENGTH, LATENT_DIM), device=device)
        fake_labels = real_labels  # Use same labels for conditional generation
        fake_data = generator(noise, fake_labels)
        fake_validity = discriminator(fake_data.detach(), fake_labels)
        d_fake_loss = adversarial_loss(fake_validity, fake)
        correct_fake += ((fake_validity < 0.5).float() == fake).sum().item()

        # Total discriminator loss
        d_loss = (d_real_loss + d_fake_loss) / 2
        d_loss.backward()
        optimizer_D.step()

        # -----------------
        #  Train Generator
        # -----------------
        optimizer_G.zero_grad()

        # Generate fake data
        gen_data = generator(noise, fake_labels)
        validity = discriminator(gen_data, fake_labels)
        g_loss = adversarial_loss(validity, valid)

        g_loss.backward()
        optimizer_G.step()

    # Calculate accuracy
    d_accuracy = (correct_real + correct_fake) / (2 * total_samples)

    # Print progress
    print(f"[Epoch {epoch}/{NUM_EPOCHS}] D loss: {d_loss.item():.4f}, G loss: {g_loss.item():.4f}, D accuracy: {d_accuracy:.4f}")

    # Save models
    torch.save(generator.state_dict(), f'generator_epoch_{epoch}.pth')
    torch.save(discriminator.state_dict(), f'discriminator_epoch_{epoch}.pth')
```

```python
# Generate synthetic data
def generate_synthetic_samples(num_samples, activity, seq_length=SEQ_LENGTH):
    generator.eval()
    with torch.no_grad():
        # Convert activity to one-hot
        activity_code = le.transform([activity])[0]
        label = ohe.transform([[activity_code]])
        label_tensor = torch.FloatTensor(label).repeat(num_samples, 1).to(device)

        noise = torch.randn((num_samples, seq_length, LATENT_DIM), device=device)
        synthetic_data = generator(noise, label_tensor)
        return synthetic_data.cpu().numpy()

# Example: Generate 100 samples of "Walking"
synthetic_walking = generate_synthetic_samples(100, "Walking")
print("Synthetic data shape:", synthetic_walking.shape)
```
